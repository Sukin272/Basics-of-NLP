```python
import datasets

dataset=datasets.load_dataset("paws", "labeled_final")

X_train1 = [x['sentence1'] for x in dataset['train']]
X_train2 = [x['sentence2'] for x in dataset['train']]
y_train = [x['label'] for x in dataset['train']]

X_test1 = [x['sentence1'] for x in dataset['test']]
X_test2 = [x['sentence2'] for x in dataset['test']]
y_test = [x['label'] for x in dataset['test']]
```

```python
from sentence_transformers import SentenceTransformer

# model_id = "bert-base-nli-mean-tokens"
model_id="all-mpnet-base-v2"
model = SentenceTransformer(model_id)
```

```python
from sentence_transformers import InputExample

numData = len(X_train1)
train_examples = [InputExample(texts=[X_train1[i], X_train2[i]], label=y_train[i]) for i in range(int(numData))]

from torch.utils.data import DataLoader

train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=3)
```

```python
from sentence_transformers import losses

loss = losses.ContrastiveLoss(model=model)
```

```python
from sentence_transformers import util

count=0
threshhold=0.9
for i in range(len(X_test1)):
    emb1 = model.encode(X_test1[i])
    emb2 = model.encode(X_test2[i])
    sim = util.dot_score(emb1, emb2)
    print(sim,"Label:",y_test[i])
    if sim > threshhold:
        if y_test[i]==1:
            count+=1
    else:
        if y_test[i]==0:
            count+=1
    print("Accuracy:",count/(i+1))
    print(i,'/',len(X_test1))
    print()
```
