```python
op=[]

with open("threshholdFinder.txt",'r') as f:
    for line in f:
        line=line.strip('\n').strip('[').strip(']')
        a=line.split(',')
        op.append([float(a[0]),int(a[1])])
print(op)
```

```python
def check(m):
    count=0
    for i in range(len(op)):
        if m<op[i][0] and op[i][1] == 0:
            count+=1
        elif m>op[i][0] and op[i][1] == 1:
            count+=1
    return count/len(op)
            
```

```python
start = 0
end = 1
for i in range(0,1000):
    m1 = start + (end-start)/3
    m2 = end - (end-start)/3
    val1 = check(m1)
    val2 = check(m2)
    if (val1 > val2):
        start = m1
    else:
        end = m2
print(start)
```
